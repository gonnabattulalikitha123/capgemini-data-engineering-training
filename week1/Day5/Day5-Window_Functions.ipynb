```sql
select emp_name, salary,
ROW_NUMBER()
over(order by salary desc) as row_number
from employeees;
```

```sql
select emp_name, department,
ROW_NUMBER()
over(partition by department order by salary desc) as row_number
from employeees;
```

```sql
select emp_name, join_date,
ROW_NUMBER()
over(order by join_date desc) as row_number
from employeees;
```

```sql
select department, join_date,
ROW_NUMBER()
over(partition by department order by salary asc) as row_number
from employeees;
```

###Rank


```sql
select emp_name, salary,
RANK()
over(order by salary desc) as rank
from employeees;
```

```sql
select emp_name, department,
RANK()
over(partition by department order by salary desc) as RANK
from employeees;
```

```sql
select emp_name, join_date,
Rank()
over(order by join_date desc) as rank
from employeees;
```

```sql
select emp_name,department,salary,
Rank()
over(partition by department order by salary ) as rank
from employeees;
```

```sql
select emp_name,
rank() over(order by emp_name) as rank
from employeees
```

###DENSE_RANK()

```sql
select emp_name,salary,
dense_rank() over(order by salary desc)as dense_rank
from employeees
```

```sql
select emp_name,department,
dense_rank() over(partition by department order by salary desc)as dense_rank
from employeees
```

```sql
select emp_name, join_date,
dense_Rank()
over(order by join_date ) as dense_rank
from employeees;
```

```sql
select emp_name,salary,
dense_rank() over(order by salary )as dense_rank
from employeees
```

```sql
select emp_name, join_date,
dense_Rank()
over(partition by department order by join_date ) as dense_rank
from employeees;
```
